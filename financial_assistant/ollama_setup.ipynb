```python
from langchain_core.output_parsers import StrOutputParser
from langchain_community.chat_models import ChatOllama
from langchain_core.output_parsers import JsonOutputParser
from langchain.prompts import PromptTemplate

llm = ChatOllama(model="llama3.1:8b-instruct-q4_0", temperature=0)
prompt = PromptTemplate(
    template="""You are a sentiment classifier\n
    Here is the text: \n\n {text} \n\n
    Answer whether it conveys positive or negative or neutral sentiment.\n""",
    input_variables=["text"],
)
```

```python
main_chain = prompt | llm
```

```python
main_chain.invoke({"text":"hi"}).content
```

```python
prompt = PromptTemplate(
    template="""You are a sentiment classifier\n
    Here is the text: \n\n {text} \n\n
    Answer whether it conveys positive or negative or neutral sentiment.\n
    Give a score 'positive', 'negative or 'neutral based on the text.\n
    Provide the score in JSON format with a single key 'sentiment' and no preamble or explanation.""",
    input_variables=["text"],
)

main_chain = prompt | llm | JsonOutputParser()
```

```python
answer = main_chain.invoke({"text":":("})
answer
```

```python
answer['sentiment']
```

```python
prompt2 = PromptTemplate(
    template="""Explain {text}""",
    input_variables=["text"],
)
```

```python
main_chain = prompt | llm | prompt2 | llm | StrOutputParser()
```
